```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the first empirical CDF
data1 = [67, 68, 69, 70, 70, 71, 72, 72, 73, 73]
sorted_data1 = np.sort(data1)
cdf_values1 = np.arange(1, len(sorted_data1) + 1) / len(sorted_data1)

# Data for the second empirical CDF
data2 = [59, 60, 63, 64, 64, 64]
sorted_data2 = np.sort(data2)
cdf_values2 = np.arange(1, len(sorted_data2) + 1) / len(sorted_data2)

# Plot both empirical CDFs
plt.figure(figsize=(8, 6))

# First empirical CDF
plt.step(sorted_data1, cdf_values1, where="post", label="First CDF (10 points)", color='blue')
plt.scatter(sorted_data1, cdf_values1, color='blue')

# Second empirical CDF
plt.step(sorted_data2, cdf_values2, where="post", label="Second CDF (6 points)", color='green')
plt.scatter(sorted_data2, cdf_values2, color='green')

# Add labels and title
plt.title('Comparison of Two Empirical CDFs')
plt.xlabel('Data Points')
plt.ylabel('CDF')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

# Data for the first empirical CDF
data1 = [67, 68, 69, 70, 70, 71, 72, 72, 73, 73]
sorted_data1 = np.sort(data1)
cdf_values1 = np.arange(1, len(sorted_data1) + 1) / len(sorted_data1)

# Data for the second empirical CDF
data2 = [59, 60, 63, 64, 64, 64]
sorted_data2 = np.sort(data2)
cdf_values2 = np.arange(1, len(sorted_data2) + 1) / len(sorted_data2)

# Create a common grid using the union of both sorted datasets
common_grid = np.union1d(sorted_data1, sorted_data2)

# Interpolate the CDF values onto this common grid
cdf1_interp = np.interp(common_grid, sorted_data1, cdf_values1, left=0, right=1)
cdf2_interp = np.interp(common_grid, sorted_data2, cdf_values2, left=0, right=1)

# Calculate the supremum norm (maximum absolute difference between the two CDFs)
supremum_norm = np.max(np.abs(cdf1_interp - cdf2_interp))

print(f'Supremum Norm Difference: {supremum_norm}')
```
